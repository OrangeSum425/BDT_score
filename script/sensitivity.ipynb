```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
base_dir = "E:/data/sensitivity/"
plots_dir = "E:/result/sensitivity/"

df_run1 = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_run1.csv")
df_run3 = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_run3.csv")
df_all = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_all_runs.csv")
```

```python
df_bdt = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_1.0_all_runs_full_uncert.csv")
plt.figure(figsize=(17,14),dpi=300)

plt.fill_between(df_bdt['mass'],df_bdt['two_sig_down'], df_bdt['two_sig_up'], color = 'yellow', alpha=0.6 , label = r'2 $\sigma$')
plt.fill_between(df_bdt['mass'],df_bdt['one_sig_down'], df_bdt['one_sig_up'], color = 'Green', alpha=0.9 , label = r'1 $\sigma$')
plt.plot(df_bdt['mass'],df_bdt['epsilon_squared'],color='red',label='BDT',linestyle='--',linewidth=3.0)
plt.yscale('log')
plt.xscale('log')
plt.xticks(size=25)
plt.yticks(size=25)
plt.xlabel(r'$M_{A^{\prime}}$[GeV]',size=30, labelpad=20)
plt.ylabel(r'$\epsilon^2$',size=30, labelpad=20)
plt.minorticks_off()
plt.legend(fontsize=20)
plt.xlim(0.01,0.1)
plt.savefig(plots_dir + "sensitivity_bdt_full_uncertainty_with_bands.png")
plt.show()
```

```python
df = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_1.0_all_runs_full_uncert.csv")
df_01 = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_1.0_all_runs_flat_uncert.csv")
plt.figure(figsize=(17,14),dpi=300)
plt.plot(df['mass'],df['epsilon_squared'],color='red',label='full background uncertainty')
plt.plot(df_01['mass'],df_01['epsilon_squared'],color='black',label='10% background flat uncertainty')
plt.yscale('log')
plt.xscale('log')
plt.xticks(size=25)
plt.yticks(size=25)
plt.xlabel(r'$M_{A^{\prime}}$[GeV]',size=30, labelpad=20)
plt.ylabel(r'$\epsilon^2$',size=30, labelpad=20)
plt.minorticks_off()
plt.legend(fontsize=20)
plt.xlim(0.01,0.1)
plt.savefig(plots_dir + "flat_vs_full_uncertainty.png")
plt.show()
```
